```python
# Normalizing the data
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
air_norm = norm_func(air)
```

```python
air_norm
```

# Agglomerative clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
# creating dendrogram
dendrogram=sch.dendrogram(sch.linkage(air_norm,method='average'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(air_norm,method='centroid'))
```

```python
hc=AgglomerativeClustering(n_clusters=50,affinity='euclidean',linkage='average')
hc.fit(air_norm)
```

```python
from sklearn.preprocessing import StandardScaler
def hier(X,n_clusters):
    ss=StandardScaler()
    X=ss.fit_transform(X)
    hier= AgglomerativeClustering(n_clusters=n_clusters)
    y_pred=hier.fit_predict(X)
    plt.scatter(X[:,0],X[:,1],c=y_pred,cmap='Paired')
    plt.title("Hierarchical Clustering")
hier(air_norm,50)
```

```python
agg_cluster=hc.fit_predict(air_norm)
```

```python
air['h_clusterid']=agg_cluster
```

# Kmeans clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,60):
    cluster=KMeans(n_clusters=i,random_state=0)
    cluster.fit(air_norm)
    wcss.append(cluster.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,60),wcss)
plt.title('Elbow Chart')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# If we observr the elbow chart, we see that the curve becomes almost flat ans the ss values does not vary much,so we must
# fix the number of clusters somewhere between 40 to 60.
# Let's fix it to 50 and check.
```

```python
cluster1=KMeans(50,random_state=0)
cluster1.fit(air_norm)
cluster1.inertia_
```

```python
# Now let's try changing the random state.
cluster2=KMeans(50,random_state=65)
cluster2.fit(air_norm)
cluster2.inertia_
```

```python
cluster2=KMeans(50,random_state=40)
cluster2.fit(air_norm)
cluster2.inertia_
```

```python
cluster2=KMeans(50,random_state=25)
cluster2.fit(air_norm)
cluster2.inertia_
```

```python
# We observe that changing the random state does not vary the ss value drastically.
```

```python
cluster1.labels_
```

```python
from collections import Counter
Counter(cluster1.labels_)
```

```python
air['k_clusterid']=cluster1.labels_           #Assigning clusters to the data
```

```python
air.groupby('k_clusterid').agg(['mean']).reset_index()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
air_norm1= norm_func(air)
air_norm1
```

```python
dbscan=DBSCAN(eps=1,min_samples=5)
dbscan.fit(air_norm1)
```

```python
dbscan.labels_
```

```python
Counter(dbscan.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
score=silhouette_score(air_norm,dbscan.labels_,metric='euclidean')
score
```
