# Kmeans clustering

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
data_norm2 = norm_func(data.iloc[:,1:])
```

```python
data_norm2
```

```python
from sklearn.cluster import KMeans
```

```python
# In Kmeans clustering,we have to specify the number of clusters in the beginning.
# For selecting the number of clusters, we will first plot the elbow chart. 
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(data_norm2)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
 # From the elbow chart,we observethat the curve is almost flat after 4 and 5, so 4 or 5 clusters would be the better choice.
```

```python
wcss
```

```python
kmeans=KMeans(4,random_state=0)
kmeans.fit(data_norm2)
kmeans.inertia_
```

```python
kmeans1=KMeans(5,random_state=0)
kmeans1.fit(data_norm2)
kmeans1.inertia_
```

```python
# We want the sum of squares value to be small,so we will go with 5 clusters,however in this case 4 clusters will also be a good choice.
```
